```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
columns_to_scale = ['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest']
scaler = StandardScaler()
x_scaled = x[columns_to_scale].copy()
x_scaled[columns_to_scale] = scaler.fit_transform(x_scaled)

x_final = pd.concat([x[['type']], x_scaled], axis=1)
```

```python
print(x_final.shape)
print(y.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_final, y, test_size=0.3, random_state=42)
```

```python
clf = MultiOutputClassifier(RandomForestClassifier(n_estimators=10, random_state=42))
clf.fit(x_train, y_train)
```

```python
clf.score(x_test,y_test)*100
```

```python
clf.score(x_train,y_train)*100
```

```python
sample_input = pd.DataFrame({
    'type': [1],
    'amount': [181.00],
    'oldbalanceOrg': [181.0],
    'newbalanceOrig': [0.0],
    'oldbalanceDest': [0.0],
    'newbalanceDest': [0.0]
})

# Ensure sample_input has the same columns and order as x_train
sample_input = sample_input[x_train.columns]

# Make predictions
try:
    predictions = clf.predict(sample_input)
    print("Predictions:")
    print(predictions)
except Exception as e:
    print("Error during prediction:", e)
```
